```python
import numpy as np
import pandas as pd

df = pd.read_csv('weatherHistory_v1.csv')
data_series = pd.Series(data=df.iloc[:,0])
data_series.info()
```

```python
# sql: select * from data_series

result = data_series[:]
print(result.shape)
```

```python
# select * from data_series where value > 0
result = data_series[ data_series>0 ]
print(result.shape)
```

```python
# select * from data_series where value > 10 and value<20
result = data_series[ (data_series>10) & (data_series<20) ]
print(result.shape)
```
